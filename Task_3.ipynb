```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('synthetic_dataset.csv')
```

```python
plot_data  = df['Price'].dropna().head(50).reset_index(drop=True)
```

```python
plt.figure(figsize=(10,6))
plt.plot(plot_data.index, plot_data.values,marker='o',linestyle='-')
plt.title('Price trend')
plt.xlabel('item sequence')
plt.ylabel('price value')

plt.show()
```

```python
category_counts =df['Category'].value_counts()
plt.figure(figsize=(10,6))
category_counts.plot(kind='bar',color='skyblue')
plt.title('count of item per category')
plt.xlabel('category')
plt.ylabel('number of item')
plt.show()
```

```python
category_counts =df['Category'].value_counts()
plt.figure(figsize=(10,6))
category_counts.plot(kind='bar',color='skyblue')
plt.title('count of item per category')
plt.xlabel('category')
plt.ylabel('number of item')
plt.xticks(rotation=0)
plt.show()
```

```python
plot_data = df[['Price','Rating']].dropna()
plt.scatter(plot_data['Price'],plot_data['Rating'],color='purple',edgecolor='w',alpha=0.4)
plt.title('Rating vs price')
plt.xlabel('Price')
plt.ylabel('Rating')
plt.grid(True,linestyle='--',alpha=0.7)
plt.show()
plt.savegrid('scatter.png')
```

```python
correlation = plot_data['Price'].corr(plot_data['Rating'])
print(correlation)
```

```python
correlation = plot_data['Rating'].corr(plot_data['Price'])
print(correlation)
```

```python
import seaborn as sns
```

```python
box_data = df[['Category','Price']].dropna()
categories = sorted(box_data['Category'].unique())

data_to_plot = [box_data[box_data['Category']==cat]['Price'] for cat in categories]

plt.boxplot(data_to_plot,
           labels=categories,
           patch_artist=True,
           boxprops =dict(facecolor='lightblue', color='blue'),
           medianprops = dict(color='red'))

plt.title('price distrubtion by category')
plt.xlabel('category')
plt.ylabel('Price')
plt.grid(axis='y',linestyle='--')
plt.show()
```

```python
median_price = box_data.groupby('Category')['Price'].median()
print(median_price)
```

```python
print(box_data.describe())
```

```python
stat_price = box_data.groupby('Category')['Price'].describe()
print(stat_price)
```

```python
stock_count = df['Stock'].value_counts()
plt.figure(figsize=(10,7))
plt.pie(stock_count,labels=stock_count.index,startangle=140,colors=['red','green'])
plt.title(' in stock vs out of stock')
plt.axis('equal')
```
