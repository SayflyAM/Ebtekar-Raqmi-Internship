```python
import pandas as pd
```

```python
file = 'synthetic_dataset.csv'
```

```python
df = pd.read_csv(file)
```

```python
df.head()
```

```python
print(df.shape)
```

```python
print(df.isnull().sum())

```

```python
print(df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
print(df.shape)
```

```python
df["Category"] = df["Category"].fillna("Unknown")
```

```python
df.dropna(subset =['Price'],inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df['Rating'] = df['Rating'].fillna(df['Rating'].mean())
```

```python
print(df.isnull().sum())
```

```python
df['Stock'].mode()
```

```python
df['Stock'].mode()[0]
```

```python
df['Stock'] = df['Stock'].fillna('In Stock')
```

```python
df['Discount'] = df['Discount'].fillna(0)
```

```python
print(df.shape)
```

```python
print(df.isnull().sum())
```
