```python
import kagglehub
```

```python

# Download latest version
path = kagglehub.dataset_download("himelsarder/retail-product-dataset-with-missing-values")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
import os

# List files in the downloaded directory
files = os.listdir(path)
print("Files in the dataset directory:", files)

# Assuming the data is in a CSV file, let's find the first one
csv_file = None
for f in files:
    if f.endswith('.csv'):
        csv_file = f
        break

if csv_file:
    full_file_path = os.path.join(path, csv_file)
    df = pd.read_csv(full_file_path)
    print("DataFrame loaded successfully. First 5 rows:")
    print(df.head())
else:
    print("No CSV file found in the directory.")
```

```python
df = pd.read_csv(full_file_path)
```
